```python
import cv2
import urllib.request
import numpy as np
from matplotlib import pyplot as plt
```

```python
response = urllib.request.urlopen("https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fwallup.net%2Fwp-content%2Fuploads%2F2016%2F03%2F10%2F343179-landscape-nature.jpg")
img = cv2.imdecode(np.array(bytearray(response.read()), dtype=np.uint8), -1)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
imgcinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
imgsuave = cv2.GaussianBlur(imgcinza, (7, 7), 0)
(T, imgthreshbinary) = cv2.threshold(imgsuave, 160, 255, cv2.THRESH_BINARY)
(T, imgthreshbinaryinv) = cv2.threshold(imgsuave, 160, 255, cv2.THRESH_BINARY_INV)
imgbitwise = cv2.bitwise_and(img, img, mask = imgthreshbinaryinv)
```
